```python
import string
from collections import Counter
import os
import io
import re
import numpy as np

b= os.chdir(r'C:\Users\Janani\Python\Desktop\Cluster Computing\Assignment')
print(os.getcwd())  

##to avoid certain words
stop_words = io.open("excluded.txt").read()
print(stop_words)

final_words=[]

def exclude_words(entire_words_in_sentence):
       entire_words_in_sentence = str(entire_words_in_sentence).replace("\t",'').strip()  
       #print(entire_words_in_sentence)
       entire_words_in_sentence = entire_words_in_sentence.translate(str.maketrans('', '', string.punctuation))
       #print(entire_words_in_sentence)        
       words = entire_words_in_sentence.split()
       #print(words)
       for word in words:
         #print('hi2',word,type(word))
         if word not in stop_words:
            #print('hi3',word)
            final_words.append(word)

       
def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    ##text = re.sub('\w*\d\w*', '', text)
    return text
        
items_list =[]  
revs_list =[]

restaurant0_list = []
restaurant1_list = []
product0_list = []
product1_list = []
movie0_list = []
movie1_list = []

result = []

with open('Basic_File.txt', 'r') as file:
    for line in file:
        clear_line = line.replace("\n", '').replace(",", '').replace("'", '').strip()
        ### Joining the item along with the sentiment values from the last character in every line 
        ### to treat them as a key value for the further validation.
        clear_line = clean_text(clear_line)
        #print(clear_line)  
        clear_line = clear_line.replace(clear_line.split()[0],(clear_line.split()[0]+clear_line.split()[-1])+': ')
        #print(clear_line)         
        clear_line1 = clear_line[:-1]
        #print(clear_line1)        
        items,revs = clear_line1.split(": 	")
        items_list.append(items)
        revs_list.append(revs)

        if items == 'restaurant0':
           final_words=[] 
           revs = revs.replace("\t",'').strip() 
           #print('hi',revs)
           exclude_words(revs)
           restaurant0_list.append(final_words)

        if items == 'restaurant1':
           final_words=[]
           revs = revs.replace("\t",'').strip()
           exclude_words(str(revs))
           restaurant1_list.append(final_words)
        
        if items == 'product0':
          final_words=[]  
          revs = revs.replace("\t",'').strip() 
          exclude_words(str(revs))
          product0_list.append(final_words)        
        

        if items == 'product1':
          final_words=[]  
          revs = revs.replace("\t",'').strip() 
          exclude_words(str(revs))
          product1_list.append(final_words)        

        if items == 'movie0':
          final_words=[]  
          revs= revs.replace('movie0:','movie')
          revs = revs.replace("\t",'').strip() 
          exclude_words(str(revs))
          movie0_list.append(final_words)        
        

        if items == 'movie1':
          final_words=[]
          revs = revs.replace("\t",'').strip() 
          exclude_words(str(revs))
          movie1_list.append(final_words)        

        

restaurant0_list.insert(0,"Restaurant0 : ")
restaurant1_list.insert(0,"Restaurant1 : ")
product0_list.insert(0,"Product0 : ")
product1_list.insert(0,"Product1 : ")
movie0_list.insert(0,"Movie0 : ")
movie1_list.insert(0,"Movie1 : ")

items_list = np.unique(items_list)        
print("Unique values ",items_list)

 
print('Displaying the Restaurant 0 reviews: ',restaurant0_list,end="\n")
print("\n")
print('Displaying the Restaurant 1 reviews: ',restaurant1_list,end="\n")
print("\n")
print('Displaying the Products 0 reviews: ',product0_list,end="\n")
print("\n")
print('Displaying the Products 1 reviews: ',product1_list,end="\n")
print("\n")
print('Displaying the Movie 0 reviews: ',movie0_list,end="\n")
print("\n")
print('Displaying the Movie 1 reviews: ',movie1_list,end="\n")
```
